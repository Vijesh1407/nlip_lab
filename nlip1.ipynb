```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

text="Machine learning is very useful in data science, deep learning and it helps in prediction and analysis."

print("Original Text: ",text)

#Tokenization
tokens=word_tokenize(text)
print("Tokens: ",tokens)

#Punctuatuon Removal
no_punctuation=[]
for word in tokens:
    if word not in string.punctuation:
        no_punctuation.append(word)
print("No punctuation: ",no_punctuation)

#Stop word Removal
filtered_words=[]
stop_words=stopwords.words("english")
for word in no_punctuation:
    if word not in stop_words:
        filtered_words.append(word)
print("No Stop words: ",filtered_words)

#Topical Words
word_count=Counter(filtered_words)
print("Topical Words: ",word_count.most_common(3))
```
