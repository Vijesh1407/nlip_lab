```python
dictionary={
    "hello":"Vanakkam",
    "how":"Epadi",
    "are":"Irukinga",
    "you":"Neenga"
}
text=input("Enter text to translate: ").lower()

words=text.split()
trans=[]

for word in words:
    if word in dictionary:
        trans.append(dictionary[word])
    else:
        trans.append(word)
ans=" ".join(trans)
print("Tanslated text: ",ans)
```
