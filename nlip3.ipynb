```python
print("Simple Chatbot")
print("Type 'bye' to exit")

while True:
    user=input("You: ").lower()
    if user=="bye":
        print("Bot: Goodbye! Have a nice day")
        break
    elif "hi" in user or "hello" in user:
        print("Bot: Hello! How can I help you?")
    elif "name" in user:
        print("Bot: I am a simple chatbot")
    elif "machine learning" in user or "ml" in user:
        print("Bot: Machine learning is a part of Artificial intelligence")
    elif "help" in user:
        print("Bot: I can answer simple questions")
    else:
        print("Bot: Sorry, I don't understand")
```
