```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("image.jpg")

# Convert BGR to RGB (OpenCV loads BGR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Low Pass Filter (Blur)
low_pass = cv2.GaussianBlur(img_rgb, (7,7), 0)

# High Pass Filter
kernel = np.array([
    [-1,-1,-1],
    [-1, 8,-1],
    [-1,-1,-1]
])

high_pass = cv2.filter2D(img_rgb, -1, kernel)

# Show images
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Low Pass Filter")
plt.imshow(low_pass)
plt.axis("off")

plt.subplot(1,3,3)
plt.title("High Pass Filter")
plt.imshow(high_pass)
plt.axis("off")

plt.show()
```
