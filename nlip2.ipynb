```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

documents=[
    "machine learning is useful",
    "data science uses machine learning",
    "football is a popular sport",
    "machine learning helps prediction"
]

query="machine learning"

# -------- TF Ranking --------
tf_vectorizer=CountVectorizer()
tf_matrix=tf_vectorizer.fit_transform(documents+[query])
query_tf=tf_matrix[-1]
tf_scores=(tf_matrix[:-1]*query_tf.T).toarray()

print("TF Scores")
for i,score in enumerate(tf_scores):
    print(f"Doc{i+1}: {round(score[0],2)}")

# -------- IDF --------
print("\nIDF Values")
for word,idf in zip(vectorizer.get_feature_names_out(),vectorizer.idf_):
    if word in query.split():
        print(f"{word}: {round(idf,2)}")

# -------- TF-IDF Ranking --------
vectorizer=TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(documents+[query])
query_vec=tfidf_matrix[-1]
tfidf_scores=(tfidf_matrix[:-1]*query_vec.T).toarray()

print("\nTF-IDF Scores")
for i,score in enumerate(tfidf_scores):
    print(f"Doc{i+1}: {round(score[0],2)}")

# -------- Cosine Similarity --------
similarity=cosine_similarity(tfidf_matrix[-1],tfidf_matrix[:-1])
scores=similarity[0]

print("\nCosine Similarity Scores")
for i,score in enumerate(scores):
    print(f"Doc{i+1}: {round(score,2)}")

# -------- PMI / PPMI --------
words=query.split()

cooccur=sum(1 for d in documents if all(w in d for w in words))

p_w1=sum(1 for d in documents if words[0] in d)/len(documents)
p_w2=sum(1 for d in documents if words[1] in d)/len(documents)
p_w1w2=cooccur/len(documents)

PMI=np.log2(p_w1w2/(p_w1*p_w2)) if p_w1w2>0 else 0
PPMI=max(0,PMI)

print("\nPMI:",round(PMI,2))
print("PPMI:",round(PPMI,2))
```
