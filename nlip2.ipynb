```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents=[
    "machine learning is useful",
    "data science uses machine learning",
    "football is a popular sport",
    "machine learning helps prediction"
]

query="machine learning"

tf_vectorizer=CountVectorizer()
tf_matrix=tf_vectorizer.fit_transform(documents+[query])

query_tf=tf_matrix[-1]
tf_scores=(tf_matrix[:-1]*query_tf.T).toarray()
tf_rank=sorted(
    list(enumerate(tf_scores)),
    key=lambda x:x[1],
    reverse=True
)

for idx,score in tf_rank:
    print("Document: ",documents[idx])
    print("Score: ",score[0])
print()

vectorizer=TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(documents+[query])

query_vec=tfidf_matrix[-1]
tfidf_scores=(tfidf_matrix[:-1]*query_vec.T).toarray()
tfidf_rank=sorted(
    list(enumerate(tfidf_scores)),
    key=lambda x:x[1],
    reverse=True
)

for idx,score in tfidf_rank:
    print("Document: ",documents[idx])
    print("Score: ",round(score[0],3))
print()

similarity=cosine_similarity(tfidf_matrix[-1],tfidf_matrix[:-1])
scores=similarity[0]

rank=sorted(
    list(enumerate(scores)),
    key=lambda x:x[1],
    reverse=True
)

for idx,score in rank:
    print("Document: ",documents[idx])
    print("Score: ",round(score,2))
```
